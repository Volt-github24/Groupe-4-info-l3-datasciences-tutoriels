Ce bout de code python nous permet de creer un jeu de donnees artificiel concernant la temperature, l'humidite et le vent dans 12 villes d' Allemagne

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import truncnorm
def truncated_normal(mean=0, sd=1, low=0, upp=10, type=int):
    return truncnorm( (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd )

def truncated_normal_floats(mean=0, sd=1, low=0, upp=10, num=100):
    res = truncated_normal(mean=mean, sd=sd, low=low, upp=upp)
    return res.rvs(num)

def truncated_normal_ints(mean=0, sd=1, low=0, upp=10, num=100):
    res = truncated_normal(mean=mean, sd=sd, low=low, upp=upp)
    return res.rvs(num).astype(np.uint8)

# nombre d'item pour chaque classe de fleur
number_of_items_per_class = [190, 205, 230, 170]

flowers = {}
# flos Pythonem:
# On cree le tableau flowers["flos_pythonem"] en empilant les tableaux reds,greens,blues et calyx_dia sous frorme de colonne
#On procede de maniere analogue pour la creation des tableaux flowers["flos_java"],flowers["flos_margarita"],flowers["flos_artificialis"] ci-dessous
number_of_items = number_of_items_per_class[0]
reds = truncated_normal_ints(mean=254, sd=18, low=235, upp=256,num=number_of_items)
greens = truncated_normal_ints(mean=107, sd=11, low=88, upp=127,num=number_of_items)
blues = truncated_normal_ints(mean=0, sd=15, low=0, upp=20,num=number_of_items)
calyx_dia = truncated_normal_floats(3.8, 0.3, 3.4, 4.2,num=number_of_items)
data = np.column_stack((reds, greens, blues, calyx_dia))
flowers["flos_pythonem"] = data

# flos Java:
number_of_items = number_of_items_per_class[1]
reds = truncated_normal_ints(mean=245, sd=17, low=226, upp=256,num=number_of_items)
greens = truncated_normal_ints(mean=107, sd=11, low=88, upp=127,num=number_of_items)
blues = truncated_normal_ints(mean=0, sd=10, low=0, upp=20,num=number_of_items)
calyx_dia = truncated_normal_floats(3.3, 0.3, 3.0, 3.5,num=number_of_items)
data = np.column_stack((reds, greens, blues, calyx_dia))
flowers["flos_java"] = data

# flos margarita:
number_of_items = number_of_items_per_class[2]
reds = truncated_normal_ints(mean=206, sd=17, low=175, upp=238,num=number_of_items)
greens = truncated_normal_ints(mean=99, sd=14, low=80, upp=120,num=number_of_items)
blues = truncated_normal_ints(mean=1, sd=5, low=0, upp=12,num=number_of_items)
calyx_dia = truncated_normal_floats(4.1, 0.3, 3.8, 4.4,num=number_of_items)
data = np.column_stack((reds, greens, blues, calyx_dia))
flowers["flos_margarita"] = data

# flos artificialis:
number_of_items = number_of_items_per_class[3]
reds = truncated_normal_ints(mean=255, sd=8, low=2245, upp=2255,num=number_of_items)
greens = truncated_normal_ints(mean=254, sd=10, low=240, upp=255,num=number_of_items)
blues = truncated_normal_ints(mean=101, sd=5, low=90, upp=112,num=number_of_items)
calyx_dia = truncated_normal_floats(2.9, 0.4, 2.4, 3.5,num=number_of_items)
data = np.column_stack((reds, greens, blues, calyx_dia)) # On empile les tableaux reds,greens,blues et calyx_dia sous frorme de colonne pour creer un seul tableau 
flowers["flos_artificialis"] = data

# On join les tableaux flowers["flos_pythonem"],flowers["flos_java"],flowers["flos_margarita"],flowers["flos_artificialis"] que l'on stocke dans data 
data = np.concatenate((flowers["flos_pythonem"],
                      flowers["flos_java"],
                      flowers["flos_margarita"],
                      flowers["flos_artificialis"]
                      ), axis=0) 


target = np.zeros(sum(number_of_items_per_class)) 
previous_end = 0
for i in range(1, 5):
  num = number_of_items_per_class[i-1]
  beg = previous_end
  target[beg: beg + num] += i
  previous_end = beg + num
conc_data = np.concatenate((data, target.reshape(target.shape[0],1)),axis=1)
np.savetxt("strange_flowers.txt", conc_data, fmt="%2.2f",)

# Representation des nuages de points pour chaque paire formee a partir des attributs des fleurs(red,green,blue,calyx) 
import matplotlib.pyplot as plt
target_names = list(flowers.keys())
feature_names = ['red', 'green', 'blue', 'calyx']
n = 4
fig, ax = plt.subplots(n, n, figsize=(16, 16))
colors = ['blue', 'red', 'green', 'yellow']
for x in range(n):
  for y in range(n):
    xname = feature_names[x]
    yname = feature_names[y]
    for color_ind in range(len(target_names)):
      ax[x, y].scatter(data[target==color_ind, x],
                      data[target==color_ind, y],
                      label=target_names[color_ind],
                      c=colors[color_ind])
    ax[x, y].set_xlabel(xname)
    ax[x, y].set_ylabel(yname)
    ax[x, y].legend(loc='upper left')

plt.show() # Affichage des nuages de points
```

Maintenant essayons de renvoyer les valeurs des variables du dataset dans des intervalles [a,b] precis, en utilisant la formule : 

f(x) = [[(b âˆ’ a) â‹… (x âˆ’ min)] / (max âˆ’ min)] + a

```python
min_x_new, max_x_new = 33, 88 # nouvelles valeurs maximales et minimales pour la colonne des x (le b et le a)
min_y_new, max_y_new = 12, 20 # nouvelles valeurs maximales et minimales pour la colonne des y (le b et le a)
data, labels = ds.make_moons(n_samples=100,shuffle=True, noise=0.05, random_state=None) # generation des donnees comme precedement(150 echantillons)
min_x, min_y = np.ndarray.min(data[:,0]), np.ndarray.min(data[:,1]) # on recupere les anciens min des colonnes x et y
max_x, max_y = np.ndarray.max(data[:,0]), np.ndarray.max(data[:,1]) # on recupere les anciens max
data -= np.array([min_x, min_y]) # on applique la premiere partie de la formule (pour chaque colonne on enleve le min)
data *= np.array([(max_x_new - min_x_new) / (max_x - min_x), (max_y_new - min_y_new) / (max_y - min_y)]) # ensuite pour chaque colonne on fait on multiplie par (b-a)/(max-min)
data += np.array([min_x_new, min_y_new]) # on ajoute le a
data[:6] # voyons ce que les 6 premieres lignes du jeu de donnees sont devenues
```

Apres cette etape, les donnees sont toutes dans le nouvel intervalle, Maintenant definissons une fonction qui permette de le faire en prenant en parametres les donnnes, les nouvelles intervalles et en precisant s'il faut modifier le jeu de donnes original ou non

```python
def scale_data(data, new_limits, inplace=False ):
  if not inplace:
    data = data.copy()
  min_x, min_y = np.ndarray.min(data[:,0]), np.ndarray.min(data[:,1])
  max_x, max_y = np.ndarray.max(data[:,0]), np.ndarray.max(data[:,1])
  min_x_new, max_x_new = new_limits[0]
  min_y_new, max_y_new = new_limits[1]
  data -= np.array([min_x, min_y])
  data *= np.array([(max_x_new - min_x_new) / (max_x - min_x),(max_y_new - min_y_new) / (max_y - min_y)])
  data += np.array([min_x_new, min_y_new])
  if inplace:
    return None
  else:
    return data
data, labels = ds.make_moons(n_samples=100, shuffle=True, noise=0.05, random_state=None)
scale_data(data, [(1, 4), (3, 8)], inplace=True)
```

Cette fonction fait en fait le meme traitement que le precedent bout de code, sauf ici elle le fait sur une copie du jeu de donnees original
