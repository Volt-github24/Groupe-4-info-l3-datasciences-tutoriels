# TUTORIEL SUR L'ALGORITHME KNN

```python
# On affiche son index , ses carateristiques et sa classe pour chacune des fleurs d'index 0,79,99 et 101 

import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
data = iris.data
labels = iris.target
for i in [0, 79, 99, 101]:
    print(f"index: {i:3}, features: {data[i]}, label: {labels[i]}")
```

```python
# Division du jeu de donnees en jeu d'entrainement et jeu de test

np.random.seed(42)
indices = np.random.permutation(len(data))
n_training_samples = 12
learn_data = data[indices[:-n_training_samples]]
learn_labels = labels[indices[:-n_training_samples]]
test_data = data[indices[-n_training_samples:]]
test_labels = labels[indices[-n_training_samples:]]
print("The first samples of our learn set:")
print(f"{'index':7s}{'data':20s}{'label':3s}")
for i in range(5): #Affichage du jeu d'entrainment
    print(f"{i:4d} {learn_data[i]} {learn_labels[i]:3}") 
    
print("The first samples of our test set:")
print(f"{'index':7s}{'data':20s}{'label':3s}")
for i in range(5): #Affichage du jeu de test
    print(f"{i:4d} {learn_data[i]} {learn_labels[i]:3}")
```

```python
# Affichage du nuage de points en 3D de toute les observations du jeu de donnees OÃ¹ le premier axe represente le premier attribut, le second axe represente le second attribut et le 3e axe represente la somme des valeurs entre le 3e et le 4e attributs 

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
colours = ("r", "b")
X = []
for iclass in range(3):
    X.append([[], [], []])
    for i in range(len(learn_data)):
        if learn_labels[i] == iclass:
            X[iclass][0].append(learn_data[i][0])
            X[iclass][1].append(learn_data[i][1])
            X[iclass][2].append(sum(learn_data[i][2:]))
colours = ("r", "g", "y")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for iclass in range(3):
    ax.scatter(X[iclass][0], X[iclass][1], X[iclass][2], c=colours[iclass])
plt.show()
```

```python
# Fonction qui calcule la distance euclidienne 2 instances ou observations

def distance(instance1, instance2):
    return np.linalg.norm(np.subtract(instance1, instance2))
print(distance([3, 5], [1, 1]))
print(distance(learn_data[3], learn_data[44]))
```

```python
# Fonction permettant de recuperer les k plus proches proches voisins d'une observation

def get_neighbors(training_set,labels,test_instance,k,distance):
    distances = []
    for index in range(len(training_set)):
        dist = distance(test_instance, training_set[index])
        distances.append((training_set[index], dist, labels[index]))
    distances.sort(key=lambda x: x[1])
    neighbors = distances[:k]
    return neighbors

#Application de la fonction get_neighbors() pour trouver les 3 plus proches voisins pour chacune des 6 premiere observation du jeu de donnees
for i in range(5):
    neighbors = get_neighbors(learn_data,learn_labels,test_data[i],3,distance=distance)
    print("Index:",i,'\n',"Testset Data: ",test_data[i],'\n',"Testset Label: ",test_labels[i],'\n',"Neighbors:",neighbors,'\n')
```

```python
#Fonction permettant de compter le nombre d'observation par classe et retourne la classe ayant le plus grand nombre d'observation 

from collections import Counter

def vote(neighbors):
    class_counter = Counter()
    for neighbor in neighbors:
        class_counter[neighbor[2]] += 1
    return class_counter.most_common(1)[0][0]

# Appliquons la fonction vote() pour trouver la classe mojoritaire des 3 plus proches voisin de chacun des 12 premieres observations
for i in range(n_training_samples):
    neighbors = get_neighbors(learn_data,learn_labels,test_data[i],3,distance=distance)
    print("index: ", i,
        ", result of vote: ", vote(neighbors),
        ", label: ", test_labels[i],
        ", data: ", test_data[i])
```

```python
# Fonction permettant de retourner la classe majoritaires des k plus proches voisins mais egalement sa probabilite

def vote_prob(neighbors):
    class_counter = Counter()
    for neighbor in neighbors:
        class_counter[neighbor[2]] += 1
    labels, votes = zip(*class_counter.most_common())
    winner = class_counter.most_common(1)[0][0]
    votes4winner = class_counter.most_common(1)[0][1]
    return winner, votes4winner/sum(votes)

# Appliquons la fonction vote_prob() 
for i in range(n_training_samples):
    neighbors = get_neighbors(learn_data,learn_labels,test_data[i],5,distance=distance)
    print("index: ", i,
        ", vote_prob: ", vote_prob(neighbors),
        ", label: ", test_labels[i],
        ", data: ", test_data[i])
```

```python
# Fonction permettant de faire le choix de la classe en appliquant une ponderation Ã  chaque voisin fonction de sa proximitÃ© avec l'observation concernÃ©e,plus elle grande plus la ponderation est elevÃ©e et inversement

def vote_harmonic_weights(neighbors, all_results=True):
    class_counter = Counter()
    number_of_neighbors = len(neighbors)
    for index in range(number_of_neighbors):
        class_counter[neighbors[index][2]] += 1/(index+1)
    labels, votes = zip(*class_counter.most_common())
    
    winner = class_counter.most_common(1)[0][0]
    votes4winner = class_counter.most_common(1)[0][1]
    if all_results:
        total = sum(class_counter.values(), 0.0)
        for key in class_counter:
            class_counter[key] /= total
        return winner, class_counter.most_common()
    else:
        return winner, votes4winner / sum(votes)
#Appliquons cette fonction pour les 6 plus proches voisins de chacune des 12 premieres observations du jeu de donnees     
for i in range(n_training_samples):
    neighbors = get_neighbors(learn_data,learn_labels,test_data[i],6,distance=distance)
    print("index: ", i,
        ", result of vote: ",vote_harmonic_weights(neighbors,all_results=True))
```

```python
# Fonction permettant de faire le choix de la classe en considerant de la distance euclidienne entre l'observation concernÃ© et chacun de ses voisins 

def vote_distance_weights(neighbors, all_results=True):
    class_counter = Counter()
    number_of_neighbors = len(neighbors)
    for index in range(number_of_neighbors):
        dist = neighbors[index][1]
        label = neighbors[index][2]
        class_counter[label] += 1 / (dist**2 + 1)
    labels, votes = zip(*class_counter.most_common())
    
    winner = class_counter.most_common(1)[0][0]
    votes4winner = class_counter.most_common(1)[0][1]
    if all_results:
        total = sum(class_counter.values(), 0.0)
        for key in class_counter:
            class_counter[key] /= total
        return winner, class_counter.most_common()
    else:
        return winner, votes4winner / sum(votes)

#Appliquons cette fonction      
for i in range(n_training_samples):
    neighbors = get_neighbors(learn_data,learn_labels,test_data[i],6,distance=distance)
    print("index: ", i,
            ", result of vote: ",vote_distance_weights(neighbors,
            all_results=True))
```

```python
# Application de la fonction vote_distance_weights() sur un simple jeu de donnees comportant 3 classes Ã  savoir : apple, banana et orange

train_set = [(1, 2, 2),
            (-3, -2, 0),
            (1, 1, 3),
            (-3, -3, -1),
            (-3, -2, -0.5),
            (0, 0.3, 0.8),
            (-0.5, 0.6, 0.7),
            (0, 0, 0)
            ] # Observations
labels = ['apple', 'banana', 'apple',
'banana', 'apple', "orange",
'orange', 'orange'] # classes
k = 2 # nombre de voisin
for test_instance in [(0, 0, 0), (2, 2, 2),
                    (-3, -1, 0), (0, 1, 0.9),
                    (1, 1.5, 1.8), (0.9, 0.8, 1.6)]:
    neighbors = get_neighbors(train_set,labels,test_instance,k,distance=distance)
    print("vote distance weights: ",vote_distance_weights(neighbors))
```
